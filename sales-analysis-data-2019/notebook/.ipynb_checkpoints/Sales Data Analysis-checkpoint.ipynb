### Drop rows of null values

```python
null_df = yearly_data[yearly_data.isna().any(axis=1)]
null_df.head()
```

We see that entire rows are empty. So we will directly drop them

```python
yearly_data.dropna(inplace = True)
yearly_data.isna().sum()
```

```python
yearly_data.shape
```

Find 'Or' in Order Date column. (we got an error when converting month to int)

```python
temporary_df = yearly_data[yearly_data['Order Date'].str[0:2] == 'Or']
temporary_df.head()
```

These are the headers as column data. 

```python
yearly_data = yearly_data[yearly_data['Order Date'].str[0:2] != 'Or']
yearly_data.head()
```

## Augmenting data with additional columns

### Add Month Column

```python
yearly_data['Month'] = yearly_data['Order Date'].str[0:2]
# yearly_data['Month'] = pd.to_numeric(yearly_data['Month'])
yearly_data['Month'] = yearly_data['Month'].apply(pd.to_numeric)
yearly_data.head()
```

### Convert string numerics to Int ( Got error when multiplying Quantity ordered and Price Each

```python
yearly_data['Quantity Ordered'] = yearly_data['Quantity Ordered'].apply(pd.to_numeric)
yearly_data['Price Each'] = pd.to_numeric(yearly_data['Price Each'])
yearly_data.head()
```

### To get Sales Value, we need to multiply Quantity Ordered with the Price of Each Product

```python
yearly_data['Sales'] = yearly_data['Quantity Ordered'] * yearly_data['Price Each']
yearly_data.head()
```

### Add a City Column

Maybe we can extract the city value from the first , to the second , 
<br>
We can use the apply function to use string functions on the pandas series

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1] #there is a white space before and after the name so 0th index is an empty string

yearly_data['City'] = yearly_data['Purchase Address'].apply(lambda x: f'{get_city(x)} ({get_state(x)})')
# yearly_data['City'] = yearly_data['Purchase Address'].apply(lambda x: x.split(',')[1])

yearly_data.head()
```

#### What was the best month for sales? How much was earned in that month?

```python
sales_per_month = yearly_data.groupby('Month').sum()
sales_per_month
```


### Plotting the graph

```python
months = [month for month in range(1,13)]

#converting to month numbers to names
month_names = [datetime.date(1900, monthinteger, 1).strftime('%B') for monthinteger in months]

plt.figure(figsize=(15,10))
sns.set(style="whitegrid")
ax = sns.barplot(x=sales_per_month['Sales'], y=month_names )
plt.xlabel('Sales in USD ($)')
plt.ylabel('Month')
plt.show()
```

We See that December was the best month for sales. We also see that the amount of products sold and the price of those products happens to be the highest in december. This may be because of Christmas and New Year's.

January is the worst month for sales. Maybe because people are broke after December.
<br><br>
We see that there is a gradual growth in sales till april and it drops again till september. Then it spikes in October, drops down in November and then spikes again in December.
The Spike in October maybe because of Halloween but we are not sure yet.

### What time should we display advertisements to maximize likelihood of customer's buying products

```python
# Take the time from order date and see what time sell more 
yearly_data['Order Date'] = pd.to_datetime(yearly_data['Order Date']) #Take care as to_datetime takes more space in general
```

```python
yearly_data.head()
```

```python
yearly_data['Hour'] = yearly_data['Order Date'].dt.hour
yearly_data['Minute'] = yearly_data['Order Date'].dt.minute
# Is simpler to grab hr and min from datetime
yearly_data.head()
```

```python
hours = [hour for hour, df in yearly_data.groupby('Hour')]

plt.figure(figsize=(15,10))
plt.plot(hours, yearly_data.groupby(['Hour']).count(), color='green')
plt.xticks(hours)
plt.xlabel('Hours')
plt.ylabel('Number of orders')
plt.grid(True)
plt.show()
```

We should advertise a little before 11am, 12noon and then around 6pm to maximise our sales.
<br>
Those times make sense as at around 11am, people might have just finished their chores.<br>
At around the 7pm mark, people might have come back home from their offices and freshened up.
<br>
These are optimal times to show our ads.

### What products are often sold together?
Maybe see the same order id for multiple orders and then check.

```python
duplicated_orders = yearly_data[yearly_data['Order ID'].duplicated(keep=False)]
#keep false=keep all ocurances of the duplicates
duplicated_orders['Grouped'] = duplicated_orders.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
```

```python
duplicated_orders.head()
```

We see that those orders have been repeated. This is because google phone, wired headphones and wired headphones, google phone for the same order id are considered different occurances. So, we will drop all duplicate order ids and keep only one.
<br>
here, we can use keep=First

```python
duplicated_orders = duplicated_orders[['Order ID', 'Grouped']].drop_duplicates()
duplicated_orders.head()
```

```python
#God Bless StackOverflow

from itertools import combinations
from collections import Counter

count_2items = Counter()
count_3items = Counter()

for row in duplicated_orders['Grouped']:
    row_list = row.split(',')
    count_2items.update(Counter(combinations(row_list, 2)))
    count_3items.update(Counter(combinations(row_list, 3)))

    
# for row in duplicated_orders['Grouped']:
#     row_list = row.split(',')
    
# print(count_2items)
for key, value in count_2items.most_common(10):
    print(key, value)
```

```python
# print(count_3items)
for key, value in count_3items.most_common(10):
    print(key, value)
```

Be smart with promotions. So when you advertise an iPhone, put a lightening cable on sale or promote it a little more because people usually want to buy an extra lightening cable. Promote wired headphones as well.
<br>
Same goes for google phones, promote or give good deals on USB C Charging cables wired headphones with the google phone

### What product sold the most and why do you think it sold the most.

```python
product_group = yearly_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']
product_group.sum()

products = [product for product, df in product_group]

plt.figure(figsize=(15,10))
sns.set(style="whitegrid")
ax = sns.barplot(products, quantity_ordered)
# plt.bar(products, quantity_ordered, color='blue')
# plt.xticks(products, rotation = 'vertical')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.xlabel('Products')
plt.ylabel('Quantity Ordered')
plt.grid(True)
plt.show()
```

AAA Batteries sold the most followed by AA batteries.
<br>
We see that lightening cables, wired headphones and usb c charging cables sold a lot too.
<br>
It seems like these cheaper products sell a lot more than the more expensive phones and monitors and so on.
<br>
Another thing may be that Dryers and washing machines last a long time so once bought, people don't need to buy a new one for years. That's why they sell so little.
<br>
Monitors are sold but not as much either. The gaming monitor sells less than the 27 inch fhd monitor because the FHD gets the same thing done with way less of a price point. Same goes for the Ultrawide monitor.
<br><br>
###### Overlaying the products sold with their prices to confirm our hypothesis.


```python
prices = yearly_data.groupby('Product').mean()['Price Each']

# plt.figure(figsize=(15,10))
fig, ax1 = plt.subplots()

fig.set_figheight(10)
fig.set_figwidth(15)

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('QQuantity Ordered', color='g')
ax2.set_ylabel('Price in USD ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8, fontsize=14)

plt.show()
```

We accept our hypothesis that the less expensive AA and AAA batteries sell more.
<br>
We notice that LG Dryers and Washing machines are pretty expensive and so are sold really less.
<br>
However, even tho the macbook pro and thinkpad are really expensive, they sell more than the dryer and the washing machine because of the buyer base. Students and professionals buy these products. A family may have 3 working professionals with macbook pros for each member but have one dryer and one washing machine.
